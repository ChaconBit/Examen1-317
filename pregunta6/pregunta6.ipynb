```python
def fibonacci(n):
    # Verificar si n es menor que 1
    if n < 1:
        return []

    # Inicializar los vectores para los términos
    fibonacci_terms = [0] * n  # Usar una lista para los términos de Fibonacci

    # Inicializar los primeros dos términos
    fibonacci_terms[0] = 0
    if n > 1:
        fibonacci_terms[1] = 1

    # Calcular los términos restantes usando solo los términos iniciales
    for i in range(2, n):
        fibonacci_terms[i] = fibonacci_terms[i - 1] + fibonacci_terms[i - 2]

    return fibonacci_terms

def main():
    n = int(input("Ingrese el número de términos de Fibonacci que desea calcular: "))

    # Obtener la secuencia de Fibonacci
    fibonacci_sequence = fibonacci(n)

    # Imprimir los términos de Fibonacci
    print(f"Los primeros {n} términos de Fibonacci son:")
    for term in fibonacci_sequence:
        print(term, end=' ')

if __name__ == "__main__":
    main()
```
